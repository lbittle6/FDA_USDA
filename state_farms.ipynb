```python
with open('/Users/laurenbittle/jobs_resumes/TDI_challenge/project/food-enforcement-0001-of-0001.json') as data_file:
    data=json.load(data_file)
df=pd.DataFrame(data['results'])

print(df.agg({'state': 'nunique'})) # 59 unique states/regions (Canada regions incl)
print(df['state'].unique())
```

```python
dfops=pd.read_csv('/Users/laurenbittle/jobs_resumes/TDI_Challenge/project/allstates_operations_USDAquery.csv'\
                  ,header='infer'\
                  ,low_memory=False)

#Added a column of state abbreviations for joining to other 'df' later
#state_keys=state_abbrev_dict.get_state_to_abbrev()
#abbrev_keys=state_abbrev_dict.get_abbrev_to_state()
state_keys=get_state_to_abbrev()
abbrev_keys=get_abbrev_to_state()

state_col=[]#np.zeros(len(dfops['State']))
for iname in range(len(dfops['State'])):
    #state_col.append(state_abbrev_dict.get_state_info(dfops['State'][iname].title(),state_keys))
    state_col.append(get_state_info(dfops['State'][iname].title(),state_keys))
dfops['state'] = np.asarray(state_col)
dfops.info()

dfops.head()#['Data Item'].unique()
```

```python
dfops2=dfops[['Year'\
              ,'State'\
              ,'state'\
              ,'State ANSI'\
              ,'Commodity'\
              ,'Data Item'\
              ,'Domain'\
              ,'Value'\
              ,'CV (%)']]
dfops2.head()


df_dfops2_merge=pd.merge(dfops2\
                         ,df\
                         ,how='inner'\
                         ,on='state')
df_dfops2_merge.head()
```

```python
c1search=df_dfops2_merge[(df_dfops2_merge['classification'] == 'Class I')\
                         & (df_dfops2_merge['Data Item'] == 'FARM OPERATIONS - ACRES OPERATED')]
acres_I=c1search.sort_index(level='state').drop_duplicates('state').Value.str.replace(',','').astype(float) #int64

c1ops=df_dfops2_merge[(df_dfops2_merge['classification'] == 'Class I')\
                      & (df_dfops2_merge['Data Item'] == 'FARM OPERATIONS - NUMBER OF OPERATIONS')]
opers_I=c1ops.sort_index(level='state').drop_duplicates('state').Value.str.replace(',','').astype(float) #int64

state_names=c1ops.sort_index(level='state').drop_duplicates('state').state.values
```

```python
# # Understanding what types of operations run in each state
# # Lower left of plot = few small farms
# # Lower right of plot = many small farms
# # Upper left of plot = few large farms
# # Upper right of plot = many large farms

pl.ion()
pl.figure(figsize=[25,10])
pl.scatter(opers_I\
           ,acres_I\
           ,alpha=0)

for ib in range(len(state_names)):
    pl.annotate(state_names[ib]\
                ,(opers_I.values[ib],acres_I.values[ib])\
                ,horizontalalignment='center'\
                ,verticalalignment='center'\
                ,size=8\
                ,weight='ultralight')

pl.xscale('log')
pl.yscale('log')
pl.xlabel('Number of Operations',fontsize=16)
pl.ylabel('Total Acres',fontsize=16)
pl.tick_params(labelsize=16)
```
